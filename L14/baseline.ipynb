```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 1000)
```

```python
df = pd.read_csv('train.csv',index_col = 0)
df_test = pd.read_csv('test1.csv',index_col = 0)
df.sort_values('sid',inplace = True)
df_test.sort_values('sid',inplace = True)
```

```python
df.head(5)
```

```python
df_test.head(5)
```

```python
sid = df['sid']
sid2 = df_test['sid']

df.drop('sid',axis = 1,inplace = True)
df_test.drop('sid',axis = 1,inplace = True)

df.insert(0,'sid',sid)
df_test.insert(0,'sid',sid2)
```

```python
df.head()
```

```python
df_test['label'] = -1
```

```python
df_train = pd.concat([df,df_test],axis = 0)
```

```python
df_train.head()
```

```python
sid = df_train['sid']
df_train.drop('sid',axis = 1,inplace = True)
df_train.insert(0,'sid',sid)
```

```python
df_train.info()
```

```python
df_train.describe()
```

```python
df_train.columns
```

```python
columns = ['sid', 'android_id', 'apptype', 'carrier', 'dev_height', 'dev_ppi',
       'dev_width', 'label', 'lan', 'media_id', 'ntt', 'os', 'osv', 'package',
       'timestamp', 'version', 'fea_hash', 'location', 'fea1_hash',
       'cus_type']
for i in columns:
    print(i+':')
    print(len(df_train[i].value_counts()))
```

```python
df_train.isnull().sum()
```

```python
df_train.drop('lan',axis = 1,inplace = True)
```

```python
df_train.info()
```

```python
arr_columns = ['apptype', 'carrier',  'ntt', 'os', 'osv', 'package',
               'version','location','cus_type','fea_hash']
for i in arr_columns:
    df_train[i] = pd.factorize(df_train[i])[0]
```

```python
train = df_train[df_train['label']!=-1]
test = df_train[df_train['label']==-1]
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(train.drop('label',axis=1), train['label'], test_size=.2)

model = xgb.XGBClassifier(
    max_depth=8,
    n_estimators=1000,
    min_child_weight=300, 
    colsample_bytree=0.8, 
    subsample=0.8, 
    eta=0.3,    
    seed=42)
```

```python
model.fit(
    X_train, y_train,
    eval_metric='auc', eval_set=[(X_train, y_train), (X_valid, y_valid)],
    verbose=True,
    #auc10epochstop
    early_stopping_rounds=10
)
```

```python
model.save_model('baseline.model')
```

```python
prob = model.predict(test.drop('label',axis = 1))
test['label']= prob
submission = test[['sid','label']]
submission.to_csv('submission.csv',index = False)
```

```python
# feature importance
import matplotlib.pyplot as plt
print(model.feature_importances_)
plt.bar(range(len(model.feature_importances_)), model.feature_importances_)
plt.show()
```
