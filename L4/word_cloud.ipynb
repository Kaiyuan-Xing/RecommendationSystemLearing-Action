```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import jieba
```

```python
#将停用词去掉
def remove_stopword(f):
    stop_words = ['一个','公司']
    for stop_word in stop_words:
        f = f.replace(stop_word,'')
    return f
```

```python
def creat_word_cloud(f):
    print('生成词云！')
    f = remove_stopword(f)
    cut_test = jieba.cut(f)
    cut_test = ''.join(cut_test)
    wc = WordCloud(
        font_path = 'simhei.ttf',
        max_words = 100,
        width = 2000,
        height = 1200)
    wordcloud = wc.generate(cut_test)
    wordcloud.to_file('word_cloud.jpg')
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
f = open('news.txt','r')
str = f.read()
```

```python
creat_word_cloud(str)
```
