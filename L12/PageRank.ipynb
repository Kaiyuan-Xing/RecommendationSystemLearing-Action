```python
import networkx as nx
import  matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.DiGraph()
```

```python
edges = [('A','B'),('A','D'),('A','E'),('A','F'),('B','C'),('C','E'),
         ('D','A'),('D','C'),('D','E'),('E','B'),('E','C'),('F','D')]
for edge in edges:
    G.add_edge(edge[0],edge[1])
```

```python
layot = nx.spring_layout(G)
nx.draw(G,pos = layot,with_labels = True,hold = False)
plt.show()
```

```python
pr = nx.pagerank(G, alpha=1)
print("简化模型的PR值：", pr)

pr = nx.pagerank(G, alpha=0.8)
print("随机模型的PR值：", pr)
```

影响力排序：C=E>B>D>A>F

```python
#转移矩阵
a = np.array([[0,0,0,1/3,0,0],
    [1/4,0,0,0,1/2,0],
    [0,1,0,1/3,1/2,0],
    [1/4,0,0,0,0,1],
    [1/4,0,1,1/3,0,0,],
    [1/4,0,0,0,0,0]])
```

```python
b = np.array([1/6,1/6,1/6,1/6,1/6,1/6])
w = b 
```

```python
#简化模型
def work(a,w):
    for i in range(100):
        w = np.dot(a,w)
        print(w)
```

```python
work(a,w)
```

```python
#随机模型
def random_work(a,w,n):
    d = 0.85
    for i in range(100):
        w = (1-d)/n + d*np.dot(a, w)
        print(w)
```

```python
random_work(a,w,6)
```

影响力排序：C=E>B>D>A>F
